```python
# Importing necessary libraries for data analysis and visualization
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
file_path = "C:\\Users\\Hp\\Desktop\\Heart Attack)\\heart_attack_prediction_india.csv"
df = pd.read_csv(file_path)

# Displaying the first few rows and basic information about the dataset
head = df.head()
info = df.info()

print(head)
print(info)
```

```python
# Checking for missing values in the dataset
missing_values = df.isnull().sum()

# Displaying the missing values
print(missing_values)
```

```python
# Analyzing the statistical summary of the dataset
summary_stats = df.describe(include='all')

# Displaying the summary statistics
print(summary_stats)
```

```python
# Visualizing the distribution of Age in the dataset
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Visualizing the correlation matrix to understand relationships between variables
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Selecting only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Calculating the correlation matrix
correlation_matrix = numeric_df.corr()

# Displaying the correlation matrix
print(correlation_matrix)
```

```python
# Importing necessary libraries for visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Setting the size of the heatmap
plt.figure(figsize=(12, 10))

# Creating a heatmap for the correlation matrix
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={'shrink': .8})

# Adding title
plt.title('Correlation Matrix Heatmap')

# Displaying the heatmap
plt.show()
```

```python
# Extracting the correlation values for 'Heart_Attack_Risk'
heart_attack_correlations = correlation_matrix['Heart_Attack_Risk'].drop('Heart_Attack_Risk')

# Identifying the top correlations (both positive and negative)
top_correlations = heart_attack_correlations.abs().nlargest(5)

# Displaying the top correlations
print(top_correlations)
```

```python
# Importing necessary libraries for visualization
import seaborn as sns
import matplotlib.pyplot as plt

# Setting up the figure for scatter plots
plt.figure(figsize=(15, 5))

# Scatter plot for LDL_Level vs Heart_Attack_Risk
plt.subplot(1, 3, 1)
sns.scatterplot(data=df, x='LDL_Level', y='Heart_Attack_Risk')
plt.title('LDL Level vs Heart Attack Risk')
plt.xlabel('LDL Level')
plt.ylabel('Heart Attack Risk')

# Scatter plot for Alcohol_Consumption vs Heart_Attack_Risk
plt.subplot(1, 3, 2)
sns.scatterplot(data=df, x='Alcohol_Consumption', y='Heart_Attack_Risk')
plt.title('Alcohol Consumption vs Heart Attack Risk')
plt.xlabel('Alcohol Consumption')
plt.ylabel('Heart Attack Risk')

# Scatter plot for Age vs Heart_Attack_Risk
plt.subplot(1, 3, 3)
sns.scatterplot(data=df, x='Age', y='Heart_Attack_Risk')
plt.title('Age vs Heart Attack Risk')
plt.xlabel('Age')
plt.ylabel('Heart Attack Risk')

# Displaying the scatter plots
plt.tight_layout()
plt.show()
```

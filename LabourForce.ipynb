```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
file_path = "C:\\Users\\Hp\\Desktop\\Labor_Force_Participation_Rate_Employment_Rate_Unemployment_Rate_and_Underemployment_Rate_Philippines_January_2005_to_January_2025.csv"
df = pd.read_csv(file_path)

# Displaying the first few rows and the summary of the dataframe
head = df.head()
info = df.info()




# Printing the outputs
print(head)
print(info)
```

```python
df.head(20)
```

```python
df['Monthly/Quarterly']=df['Monthly/Quarterly'].replace('Annual','January')
df= df.replace('..','0')
df.head(20)
```

```python
# Checking for missing values
missing_values = df.isnull().sum()

# Descriptive statistics
description = df.describe()
print(missing_values)
print(description)
```

```python
# Checking the columns and the first few rows of the dataframe to understand its structure
columns = df.columns.tolist()
head = df.head()

# Printing the columns and the head of the dataframe
print(columns)
print(head)
```

```python
# Visualizing the data
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Labor Force Participation Rate', label='Labor Force Participation Rate')
sns.lineplot(data=df, x='Year', y='Employment Rate', label='Employment Rate')
sns.lineplot(data=df, x='Year', y='Unemployment Rate', label='Unemployment Rate')
sns.lineplot(data=df, x='Year', y='Underemployment Rate', label='Underemployment Rate')
plt.title('Labor Force and Employment Rates in the Philippines (2005-2025)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Monthly/Quarterly'], format='%Y-%B')

# Now, let's visualize the data again
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='Labor Force Participation Rate', label='Labor Force Participation Rate')
sns.lineplot(data=df, x='Date', y='Employment Rate', label='Employment Rate')
sns.lineplot(data=df, x='Date', y='Unemployment Rate', label='Unemployment Rate')
sns.lineplot(data=df, x='Date', y='Underemployment Rate', label='Underemployment Rate')
plt.title('Labor Force and Employment Rates in the Philippines (2005-2025)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Checking the unique values in the 'Monthly/Quarterly' column to identify any invalid entries
unique_months = df['Monthly/Quarterly'].unique()

# Printing the unique values
print(unique_months)
```

```python
# Filtering out rows where 'Monthly/Quarterly' is 'Annual'
df_filtered = df[df['Monthly/Quarterly'] != 'Annual']

# Creating a 'Date' column by combining 'Year' and 'Monthly/Quarterly'
df_filtered['Date'] = pd.to_datetime(df_filtered['Year'].astype(str) + '-' + df_filtered['Monthly/Quarterly'], format='%Y-%B')

# Now, let's visualize the data again
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_filtered, x='Date', y='Labor Force Participation Rate', label='Labor Force Participation Rate')
sns.lineplot(data=df_filtered, x='Date', y='Employment Rate', label='Employment Rate')
sns.lineplot(data=df_filtered, x='Date', y='Unemployment Rate', label='Unemployment Rate')
sns.lineplot(data=df_filtered, x='Date', y='Underemployment Rate', label='Underemployment Rate')
plt.title('Labor Force and Employment Rates in the Philippines (2005-2025)')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Checking for missing values in the filtered dataframe
missing_values_filtered = df_filtered.isnull().sum()

# Descriptive statistics for the filtered dataframe
description_filtered = df_filtered.describe()


# Printing the outputs
print(missing_values_filtered)
print(description_filtered)
```

```python
# Dropping non-numeric columns for correlation analysis
numeric_df = df_filtered.drop(columns=['Year', 'Monthly/Quarterly'])

# Checking for missing values in the numeric dataframe
missing_values_numeric = numeric_df.isnull().sum()

# Descriptive statistics for the numeric dataframe
description_numeric = numeric_df.describe()

# Correlation matrix
correlation_matrix_numeric = numeric_df.corr()

# Printing the outputs
print(missing_values_numeric)
print(description_numeric)
print(correlation_matrix_numeric)
```

```python
# Converting relevant columns to numeric, coercing errors to NaN
numeric_columns = ['Labor Force Participation Rate', 'Employment Rate', 'Unemployment Rate', 'Underemployment Rate']

for col in numeric_columns:
    df_filtered[col] = pd.to_numeric(df_filtered[col], errors='coerce')

# Checking for missing values in the cleaned numeric dataframe
missing_values_cleaned = df_filtered[numeric_columns].isnull().sum()

# Descriptive statistics for the cleaned numeric dataframe
description_cleaned = df_filtered[numeric_columns].describe()

# Correlation matrix for the cleaned numeric dataframe
correlation_matrix_cleaned = df_filtered[numeric_columns].corr()

# Printing the outputs
print(missing_values_cleaned)
print(description_cleaned)
print(correlation_matrix_cleaned)
```

```python
# Creating a heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_cleaned, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap of Employment Rates in the Philippines')
plt.tight_layout()
plt.show()
```

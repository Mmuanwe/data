```python
# to import the label encoder module
from sklearn import preprocessing

# to create a label encoder object
label_encoder = preprocessing.LabelEncoder()

# applying to the categorical columns
df['Brand'] = label_encoder.fit_transform(df['Brand'])
df['Model'] = label_encoder.fit_transform(df['Model'])
df['Year'] = label_encoder.fit_transform(df['Year'])
df['Country'] = label_encoder.fit_transform(df['Country'])
df['Condition'] = label_encoder.fit_transform(df['Condition'])
df['Engine_Size'] = label_encoder.fit_transform(df['Engine_Size'])
df['Horsepower'] = label_encoder.fit_transform(df['Horsepower'])
df['Torque'] = label_encoder.fit_transform(df['Torque'])
df['Top_Speed'] = label_encoder.fit_transform(df['Top_Speed'])
df['Acceleration_0_100'] = label_encoder.fit_transform(df['Acceleration_0_100'])
df['Fuel_Type'] = label_encoder.fit_transform(df['Fuel_Type'])
df['Drivetrain'] = label_encoder.fit_transform(df['Drivetrain'])
df['Transmission'] = label_encoder.fit_transform(df['Transmission'])
df['Fuel_Efficiency'] = label_encoder.fit_transform(df['Fuel_Efficiency'])
df['CO2_Emissions'] = label_encoder.fit_transform(df['CO2_Emissions'])
df['Price'] = label_encoder.fit_transform(df['Price'])
df['Mileage'] = label_encoder.fit_transform(df['Mileage'])
df['Popularity'] = label_encoder.fit_transform(df['Popularity'])
df['Safety_Rating'] = label_encoder.fit_transform(df['Safety_Rating'])
df['Number_of_Owners'] = label_encoder.fit_transform(df['Number_of_Owners'])
df['Market_Demand'] = label_encoder.fit_transform(df['Market_Demand'])
df['Insurance_Costs'] = label_encoder.fit_transform(df['Insurance_Cost'])
df['Production_Units'] = label_encoder.fit_transform(df['Production_Units'])
df['Log_Price'] = label_encoder.fit_transform(df['Log_Price'])
df['Log_Mileage'] = label_encoder.fit_transform(df['Log_Mileage'])
df['Modification'] = label_encoder.fit_transform(df['Modification'])
```

SEPERAING THE DATA SETS 

```python
# Creating a new dataframe for the independent varibles known as X
X = df.drop('Price', axis =1)
X.head()
```

```python
# Creating a new dataframe for the dependent varibles known as y
y = df['Price']
y.head()
```

```python
# To create the train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 10)
```

FEATURE SCALING

```python
# importing the Standard Scaler module
from sklearn.preprocessing import StandardScaler

# Create a scaler object
sc = StandardScaler()

# Applying it to the X_train data
X_train = sc.fit_transform(X_train)
pd.DataFrame(X_train)
```

```python
# Applying it to the X_test data
X_test = sc.fit_transform(X_test)
X_test
```

MODEL BUILDING

```python
# importing the linear regression model
from sklearn.linear_model import LinearRegression

# Creating a linear regression object
reg = LinearRegression()

# Model traing
reg.fit(X_train, y_train)
```

MODEL TESTING

```python
# To predict the price of laptops in our test dataset, using the model we just built
y_pred = reg.predict(X_test)
y_pred_table=pd.DataFrame(y_pred)
y_pred_table
                
```

```python
# To predict the price of laptops in our test dataset, using the model we just built
y_pred = reg.predict(X_test)
y_pred_table=pd.DataFrame(y_pred)
y_pred_table
              
```

```python
from sklearn import metrics
import numpy as np  
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", metrics.mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:",np.sqrt(metrics.mean_absolute_error(y_test, y_pred)) )
```

```python
# Evaluating using the r^2  method, this is the default metric for evaluating a linear model.
reg.score(X_test, y_test)
```

Linear Regression model is suitable but lets try out other models that can be suited for this DATA

```python
# Futher evaluation using the cross validation method
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cross_valid = ShuffleSplit(n_splits = 9, test_size = 0.2, random_state = 10)
cross_val_score(LinearRegression(), X, y, cv = cross_valid)
```
